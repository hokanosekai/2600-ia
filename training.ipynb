```python
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Load the latest version
df = kagglehub.dataset_load(
    KaggleDatasetAdapter.PANDAS,
    "hemanthpingali/phishing-url",
    "Training.parquet"
)
```

```python
df
```

### import pandas as pd
pd.unique(df["status"])

```python
df.columns
```

```python
X = df.drop(columns=["url","status"])
y = df["status"]
```

```python
print(X)
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder() #instancie un label encoder qui va subsituer les str en int
y_encoded = le.fit_transform(y) # apprend et transform
```

```python
y_encoded
```

```python
print(X["length_url"].min())
print(X["length_url"].max())

print(X["domain_in_title"].min())
print(X["domain_in_title"].max())
```

```python
# attention au null
df.dropna
```

```python
# train_test_split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=2600)

# chargement d'un model
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns = X.columns)

print("-"*20)
print(X_scaled["length_url"].max())
print(X_scaled["domain_in_title"].max())
print(X_scaled["length_url"].min())
print(X_scaled["domain_in_title"].min())
print("-"*20)

# fit
rdf = RandomForestClassifier()
rdf.fit(X_scaled, y_train)

X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns = X.columns)

y_pred_rdf = rdf.predict(X_test_scaled)

from sklearn.metrics import accuracy_score, f1_score, classification_report

stat = accuracy_score(y_test, y_pred_rdf)
f1 = f1_score(y_test, y_pred_rdf)
report = classification_report(y_test, y_pred_rdf)

print(stat)
print(f1)
print(report)
```

```python
import skops.io as skio

to_save = [le, scaler, rdf]
with open("model.skio", "wb") as f:
    skio.dump(to_save, f)
```
